```python
import pandas as pd
data = {
    "order_id":[1,2,3,4],
    "restaurant_name":["Mc Donalds", "Pizza Hut", "Dominos", "La Pinoz Pizza"],
    "dish_name":["French fries", "Nachos", "Pizza","Pasta"],
    "category":["Veg","Veg","Veg","Veg"],
    "price":[200,150,300,250],
    "Quantity":[2,1,2,2],
    "Order_month":["Jan","Mar","Feb","May"],
    "Rating":[3,4,5,5],
    "Payment_mode":["Cash","Online","Card","Cash"]
}
df=pd.DataFrame(data)
print(df)

```

```python
#calculating total revenue#
df["revenue"] = df["price"] * df["Quantity"]
df
```

```python
#sum of category#
df.groupby("category")["revenue"].sum()
```

```python
#sorting using groupby#
df.groupby("dish_name")["Quantity"].sum().sort_values(ascending=False)
```

```python

df.groupby("restaurant_name")["revenue"].sum().sort_values(ascending=False)
```
